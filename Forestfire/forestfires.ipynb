```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("forestfires.csv")
```

```python
data
```

```python
data.shape
```

```python
month_mapping = {'jan': 1, 'feb': 2, 'mar': 3, 'apr': 4, 'may': 5, 'jun': 6,
'jul': 7, 'aug': 8, 'sep': 9, 'oct': 10, 'nov': 11, 'dec': 12}
day_mapping = {'mon': 1, 'tue': 2, 'wed': 3, 'thu': 4, 'fri': 5, 'sat': 6, 'sun': 7}
data['month'] = data['month'].map(month_mapping)
data['day'] = data['day'].map(day_mapping)
```

```python
plt.figure(figsize=(10, 6))
sns.lineplot(data=data, x="day", y="temp", errorbar=None)
plt.title("Temperature Over Days")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(data=data, x="month", y="area", errorbar=None)
plt.title("Area Burned by Month")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x="wind", y="temp")
plt.title("Wind vs. Temperature")
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.heatmap(data.corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Heatmap of Feature Correlations")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=data, x="area", bins=30, kde=True)
plt.title("Distribution of Area Burned")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=data, x="month", y="temp")
plt.title("Temperature Distribution by Month")
plt.show()
```

```python
sns.pairplot(data[["temp","RH","wind","rain"]])
plt.suptitle("Pair plot of Selected Features", y=1.02)
plt.show()
```
